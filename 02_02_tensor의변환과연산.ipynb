```python
import numpy as np
import torch
```

```python
tensor = torch.ones(2,2)
```

```python
# tensor와 메모리 공유 
np_array = tensor.numpy()  # 참조(같은 메모리 주소 가리킴)
np_array
```

```python
# tensor 복사본
np_array2 = np.array(tensor) # 새 메모리 할당
np_array2
```

```python
tensor[0,0] = -1
```

```python
print(f'numpy() result\n{np_array}\n')

print(f'np.array() result\n{np_array2}')
```
